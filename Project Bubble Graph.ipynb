```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("cleaned_final_dataset.csv")
```

```python
data
```

```python
data_new = data.loc[1:200, ['valence', 'energy', 'danceability']]
sizes = data_new['valence'] * 1200

# Create a scatter plot with bubble sizes
plt.scatter(data_new["danceability"], data_new["energy"], s=sizes, alpha=0.20)

plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.title('Bubble Chart DANCEABILITY vs ENERGY vs VALENCE')
plt.show()
```
