```python
import pandas as pd
import numpy  as np 
```

```python
data30 = pd.read_csv("Cleaned_data.csv")

data30['Artist'] = data30['Artist'].str.lower()
data30['Track'] = data30['Track'].str.lower()
data30['Artist'] = data30['Artist'].str.strip()
data30['Track'] = data30['Track'].str.strip()
data30 = data30.dropna(subset=['Track'])
data30 = data30.dropna(subset=['Artist'])
data30.drop('Unnamed: 0', axis=1, inplace=True)
```

In this part, we deleted the other words coming after the Name of the artist because in the second dataset, there was only the Artist's name and surname, nothing more

```python
data30['Track'] = data30['Track'].apply(lambda x: x.split('(')[0].strip() if '(' in x else x)
```

```python
lowered_columns = [colname.lower() for colname in list((data30.columns))]
data30.columns = lowered_columns
```

```python
data30.rename(columns={'track':'title', 'title':'full_title'}, inplace=True)
```

```python
data30.columns
```

```python
sample = pd.read_csv('ds2.csv', nrows=10)
sample['title'] = sample['title'].str.lower()
sample['title'] = sample['title'].str.strip()
sample['artist'] = sample['artist'].str.lower()
sample['artist'] = sample['artist'].str.strip()
sample.head(10)
```

```python
merged_data  = pd.merge(data30, sample, on=['artist','title'], how='inner')
merged_data
```

We created an empty list, now the code will read by parts the second dataset and add to empty list

```python
i = 0
iterator = pd.read_csv("ds2.csv", chunksize = 100000)
type(iterator)
for chunk in iterator:


    merged_data2 = chunk
   # merged_data2.columns = cols
    merged_data2['title'] = merged_data2['title'].str.lower()
    merged_data2['title'] = merged_data2['title'].str.strip()
    merged_data2['artist'] = merged_data2['artist'].str.lower()
    merged_data2['artist'] = merged_data2['artist'].str.strip()
    df = pd.merge(data30, merged_data2,  on = ['artist', 'title'], how='inner')
    merged_data = pd.concat([merged_data, df])
    i+=1
    print(f'Iteration {i} passed successfully!')
```

```python
merged_data.to_csv('merged_dataset.csv')
```
